```
call_durations = [2, 5, None, 45, 120, 1, 300, None, 15]

def rem_miss(call_durs):
    return [x for x in call_durs if x is not None]

cleaned = rem_miss(call_durations)
print("After removing missing values:", cleaned)

def rem_invalid(calls):
    return [x for x in calls if 2 <= x <= 180]

rem_list = rem_invalid(cleaned)
print("Removal of invalid values (>=2 & <=180):", rem_list)

def avg_call(calls):
    return sum(calls) / len(calls)

avg = avg_call(rem_list)
print("Average call duration:", avg)

def long_calls(calls):
    count = 0
    total = len(calls)
    for x in calls:
        if x > 30:
            count += 1
    call_per = (count / total) * 100
    return call_per

c = long_calls(rem_list)
print(f"Percentage of calls greater than 30 mins is {c}%")
```

```
orders = [
    {"price": 120, "quantity": 2},
    {"price": -50, "quantity": 1},
    {"price": 300, "quantity": 0},
    {"price": 80, "quantity": 3}
]

def is_valid_order(order):
    return (
        isinstance(order["price"], (int, float)) and
        isinstance(order["quantity"], (int, float)) and
        order["price"] > 0 and
        order["quantity"] > 0
    )

# Test
print(is_valid_order(orders[0]))  # True
```
