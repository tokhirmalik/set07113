```
# Exercise 2: Create class Rectangle with method to compute area
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Create object
rect = Rectangle(5, 4)

# Output
print("\nExercise 2 Output:")
print("Area:", rect.area())
```

```
# Exercise 3: Create class Employee with salary
# and method to compute annual salary

class Employee:
    def __init__(self, salary):
        self.salary = salary

    def annual_salary(self):
        return self.salary * 12


# Create object
emp = Employee(2000)

# Output
print("Exercise 3 Output:")
print("Annual Salary:", emp.annual_salary())
```

```
# TASK A: Product Class

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, amount):
        self.quantity += amount

    def total_value(self):
        return self.price * self.quantity

    def apply_discount(self, percent):
        self.price -= self.price * (percent / 100)


# Create products
p1 = Product("Laptop", 1000, 5)
p2 = Product("Phone", 500, 10)
p3 = Product("Tablet", 300, 7)

products = [p1, p2, p3]

# 1️⃣ Total inventory value
total_inventory = sum(p.total_value() for p in products)
print("Total Inventory Value:", total_inventory)

# 2️⃣ Apply discount
p1.apply_discount(10)
print("Laptop price after discount:", p1.price)

# 3️⃣ Find most expensive product
most_expensive = max(products, key=lambda p: p.price)
print("Most expensive product:", most_expensive.name)
```

```
# TASK B: Shape Class
import math

class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

    def scale(self, factor):
        self.radius *= factor


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def scale(self, factor):
        self.width *= factor
        self.height *= factor

    def can_fit_inside(self, other):
        return self.width <= other.width and self.height <= other.height


# Example usage
circle = Circle(5)
rect1 = Rectangle(4, 6)
rect2 = Rectangle(10, 12)

print("Circle area:", circle.area())
print("Rectangle area:", rect1.area())
print("Can rect1 fit inside rect2?", rect1.can_fit_inside(rect2))

circle.scale(2)
print("Scaled circle radius:", circle.radius)
```

```
# TASK C: Student Class
class Student:
    def __init__(self, name):
        self.name = name
        self.courses = {}

    def add_course(self, course, grade=None):
        self.courses[course] = grade

    def average_grade(self):
        grades = [g for g in self.courses.values() if g is not None]
        if not grades:
            return 0
        return sum(grades) / len(grades)

    def incomplete_courses(self):
        return [course for course, grade in self.courses.items() if grade is None]


# Create students
s1 = Student("Ali")
s2 = Student("Vali")
s3 = Student("Hasan")

s1.add_course("Math", 90)
s1.add_course("Python", 85)

s2.add_course("Math", 60)
s2.add_course("Python", None)

s3.add_course("Math", 95)
s3.add_course("Python", 92)

students = [s1, s2, s3]

# 1️⃣ Best student
best_student = max(students, key=lambda s: s.average_grade())
print("Best student:", best_student.name)

# 2️⃣ Incomplete courses
print("Vali incomplete courses:", s2.incomplete_courses())

# 3️⃣ Top-performing students (average >= 90)
top_students = [s.name for s in students if s.average_grade() >= 90]
print("Top-performing students:", top_students)
```
